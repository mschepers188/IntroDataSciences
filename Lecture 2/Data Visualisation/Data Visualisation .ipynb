```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('1976-2016-president.csv')
```

```python
data
```

```python
data[data.state == 'Wisconsin']
```

```python
Wisconsin=data[data.state == 'Wisconsin']
```

```python
Democrats = Wisconsin[Wisconsin.party=='democrat']
```

```python
Republicans = Wisconsin[Wisconsin.party=='republican']
```

```python
plt.plot(Wisconsin.year, Wisconsin.totalvotes/ 10**6)
plt.show()
```

```python
plt.plot(Democrats.year, Democrats.candidatevotes/ 10**6)
plt.show()
```

```python
plt.plot(Republicans.year, Republicans.candidatevotes/ 10**6)
plt.show()
```

```python
plt.plot(Democrats.year, Democrats.candidatevotes/ 10**6)
plt.plot(Republicans.year, Republicans.candidatevotes/ 10**6)
plt.legend(['Democrats', 'Republicans'])
plt.xlabel('Year')
plt.ylabel('Candidate Vote')
plt.show()
```

```python
plt.bar(Democrats.year, Democrats.candidatevotes/ 10**6, color ='blue',  
        width = 1) 
plt.bar(Republicans.year, Republicans.candidatevotes/ 10**6, color ='red',  
        width = 1)
plt.legend(['Democrats', 'Republicans'])
plt.xlabel('Year')
plt.ylabel('Candidate Vote')
plt.show()
```

```python
plt.bar(Democrats.year, Democrats.candidatevotes/ 10**6, color ='blue',  
        width = 1) 
plt.bar(Republicans.year+1, Republicans.candidatevotes/ 10**6, color ='red',  
        width = 1)
plt.legend(['Democrats', 'Republicans'])
plt.xlabel('Year')
plt.ylabel('Candidate Vote')
plt.show()
```
