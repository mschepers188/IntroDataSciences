```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
source = requests.get('https://foodwishes.blogspot.com').text

soup = BeautifulSoup(source, 'lxml')
```

```python
print(soup)
```

```python
csv_file = open('FoodWishes.csv', 'w')
```

```python
csv_writer = csv.writer(csv_file)
csv_writer.writerow(['headline', 'summary', 'video_link'])
```

```python
Post= soup.find('div', class_='post hentry')
print(Post)
```

```python
Post_title= soup.find('div', class_='post hentry').h3.text
print(Post_title)
```

```python
Summary = Post.find('div', class_='post-body entry-content').text
print(Summary)
```

```python
Link= Post.find('iframe')['src']
print(Link)
```

```python
for Post in soup.find_all('div', class_='post hentry'):
    headline = Post.h3.text
    print(headline)
    
    summary = Post.find('div', class_='post-body entry-content').text
    print(summary)
    
    video_link= Post.find('iframe')['src']
    print(video_link)

    print()
    
    csv_writer.writerow([headline, summary, video_link])

csv_file.close()

   
```
